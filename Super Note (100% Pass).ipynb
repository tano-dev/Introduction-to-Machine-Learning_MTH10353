## Week 0 - Introduction to Machine Learning


### 1. Tổng quan về AI, Machine Learning và Deep Learning
* **Trí tuệ nhân tạo (Artificial Intelligence):** Là một chương trình có khả năng nhận thức, suy luận, hành động và thích nghi.
* **Máy học (Machine Learning):** Là các thuật toán có khả năng cải thiện hiệu suất khi chúng được tiếp xúc với nhiều dữ liệu hơn theo thời gian.
* **Học sâu (Deep Learning):** Là một tập con của máy học, trong đó các mạng nơ-ron nhiều lớp học hỏi từ những lượng dữ liệu khổng lồ.
### 2. Tại sao nên sử dụng Máy học?
Đối với nhiều vấn đề, rất khó để lập trình chính xác hành vi bằng tay (ví dụ như nhận diện con người, vật thể hoặc hiểu giọng nói con người).

* **Cách tiếp cận của Máy học:** Lập trình một thuật toán để tự động học từ dữ liệu hoặc từ kinh nghiệm.


* **Lý do sử dụng thuật toán học tập:**
    * Giải pháp khó có thể lập trình thủ công bằng tay (như thị giác máy tính, giọng nói).
    * Hệ thống cần thích nghi với một môi trường luôn thay đổi (như phát hiện email rác).
    * Mong muốn hệ thống có hiệu suất vượt trội hơn cả các lập trình viên con người.
    * Đảm bảo tính công bằng và riêng tư (như việc xếp hạng kết quả tìm kiếm).

### 3. Các loại hình Máy học chính
* **Học có giám sát (Supervised learning):** Hệ thống có sẵn các ví dụ đã được dán nhãn về hành vi đúng.
* **Học tăng cường (Reinforcement learning):** Hệ thống học tập nhận được một tín hiệu phần thưởng và cố gắng học cách để tối đa hóa tín hiệu phần thưởng đó.
* **Học không giám sát (Unsupervised learning):** Không có các ví dụ dán nhãn; thay vào đó, hệ thống tìm kiếm những quy luật (patterns) thú vị trong dữ liệu.
> **Ví dụ**: 
>   * Học có giám sát: Dự đoán giá nhà dựa trên các đặc điểm như diện tích, số phòng ngủ, v.v.
>   * Học tăng cường: Một robot học cách đi qua một mê cung bằng cách nhận phần thưởng khi tiến gần đến đích, và bị phạt khi đi sai đường.
>   * Học không giám sát: Phân cụm khách hàng thành các nhóm dựa trên hành vi mua sắm của họ mà không cần biết trước về các nhóm đó.
### 6. Mô hình sinh (Generative Models)
Trong mô hình sinh, mục tiêu là học được hàm phân phối của một tập dữ liệu. 
> **Ví dụ**: Hình ảnh tự nhiên, để có thể tạo ra các mẫu mới tương tự như dữ liệu gốc.
### 7. Học sâu (Deep Learning) & Công cụ nền tảng

* **Học sâu:** Bao gồm rất nhiều lớp xử lý.
* **Thị giác máy tính:** Các mạng như Faster R-CNN hay Mask R-CNN có khả năng nhận diện và phân vùng đối tượng chi tiết trong ảnh.
* **Tại sao nên dùng mạng nơ-ron?**
    * Được lấy cảm hứng từ não bộ: là bằng chứng cho thấy kiến trúc nơ-ron có thể "nhìn" và "nghe".
    * Rất hiệu quả trong nhiều ứng dụng (thị giác, văn bản, giọng nói, y tế, robotics...) và được sử dụng rộng rãi trong cả học thuật lẫn công nghiệp.

* **Công cụ phần mềm:**
    * **Xử lý mảng (NumPy):** Giúp tính toán vector hóa nhằm khai thác hiệu quả phần cứng.
    * **Các Framework (TensorFlow, PyTorch, Keras...):** Hỗ trợ vi phân tự động, biên dịch đồ thị tính toán, có sẵn thư viện thuật toán và hỗ trợ GPU.
